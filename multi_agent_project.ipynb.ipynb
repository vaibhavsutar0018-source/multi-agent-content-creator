```python
def research_agent(topic):
    return f"Key points about {topic}: benefits, challenges, applications"

def writer_agent(points):
    return f"Blog:\n{points}\nThis topic is very important in today's world."

def editor_agent(article):
    return f"Final Improved Content:\n{article}"


topic = input("Enter topic: ")

points = research_agent(topic)
article = writer_agent(points)
final = editor_agent(article)

print("\nFINAL OUTPUT:\n")
print(final)
```
